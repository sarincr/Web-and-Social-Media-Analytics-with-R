```r
i <-  0
s <-  0
cat("\n i=",i)
cat("\n Sum = ", s)
while (i < 100) {
  cat("\n i=",i)
  i <- i + 1
  cat("\n New i=",i)
  s=s+i
  cat("\n New Sum (Sum+i) = ", s)
} 
cat("\n Final Sum = ",s)
```
